```python
import pandas as pd

# Load your CSV file
df = pd.read_csv("professors_with_emails.csv")  # use your filename

# Define the cleanup function
def clean_research_areas(text):
    if pd.isna(text):
        return text
    text = text.replace(" and ", ", ")
    parts = [part.strip() for part in text.split(",")]
    unique_parts = list(dict.fromkeys(part for part in parts if part))
    return ", ".join(unique_parts)

# Apply the function
df['Cleaned Research Areas'] = df['Research Areas'].apply(clean_research_areas)

# Save to new CSV
df.to_csv("professor_cleaned.csv", index=False)
print("✅ Cleaned research areas saved to umd_cs_faculty_cleaned.csv")
```

```python
import pandas as pd
import google.generativeai as genai
import time

# Step 1: Setup Gemini
genai.configure(api_key="AIzaSyCR8TmTktb2mV4pJY4_Zt2bgticSHiCGNU")  # Replace with your actual key
model = genai.GenerativeModel('gemini-2.0-flash')

# Step 2: Prompt + Skill function
def prompt_for_skills(research_area):
    return (
        f"Based on the research areas: {research_area}, "
        f"generate a list of 7 to 10 technical skills a graduate student should have. "
        f"Return only the skill names, comma-separated, no explanation."
    )

def get_skills(research_area):
    try:
        response = model.generate_content(prompt_for_skills(research_area))
        return response.text.strip()
    except Exception as e:
        print(f"[ERROR] {e}")
        return "N/A"

# Step 3: Load only batch 1 (first 15 rows)
df = pd.read_csv("professor_cleaned.csv")
batch_df = df.iloc[0:15].copy()

# Step 4: Generate skills for this batch
skills = []
for area in batch_df["Research Areas"]:
    skills.append(get_skills(area))
    time.sleep(1.5)  # polite pause

batch_df["Skills"] = skills

# Step 5: Save batch result to CSV
batch_df.to_csv("professor_batch_1.csv", index=False)
print("✅ Batch 1 saved to professor_batch_1.csv")
```
