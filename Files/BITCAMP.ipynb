```python
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import LETTER
from reportlab.lib.units import inch
from reportlab.lib.utils import simpleSplit
import os

# Create PDF output folder
os.makedirs("resumes_pdf", exist_ok=True)

# Font settings
FONT = "Helvetica"
FONT_BOLD = "Helvetica-Bold"
LINE_SPACING = 15
SECTION_SPACING = 28
FIELD_INDENT = 120

# Wrap helper
def draw_wrapped_text(c, text, x, y, max_width, font=FONT, font_size=11, line_spacing=LINE_SPACING):
    c.setFont(font, font_size)
    lines = simpleSplit(text, font, font_size, max_width)
    for l in lines:
        c.drawString(x, y, l.strip())
        y -= line_spacing
    return y

# Process resumes
for filename in os.listdir("resumes"):
    if filename.endswith(".txt"):
        txt_path = os.path.join("resumes", filename)
        pdf_path = os.path.join("resumes_pdf", filename.replace(".txt", ".pdf"))

        with open(txt_path, "r") as f:
            lines = f.read().splitlines()

        c = canvas.Canvas(pdf_path, pagesize=LETTER)
        width, height = LETTER
        left_margin = 1.2 * inch
        y = height - 1 * inch

        # Name (centered, capitalized)
        name = lines[0].strip().upper()
        c.setFont(FONT_BOLD, 16)
        c.drawCentredString(width / 2, y, name)
        y -= SECTION_SPACING

        c.setFont(FONT, 11)

        for i in range(1, len(lines)):
            line = lines[i].strip()

            if not line:
                y -= LINE_SPACING
                continue

            if "Education:" in line:
                c.setFont(FONT_BOLD, 11)
                c.drawString(left_margin, y, "Education:")
                y -= LINE_SPACING
                continue

            if line.startswith("University of"):
                y = draw_wrapped_text(c, line, left_margin + 10, y, width - 2 * inch)
                continue

            if "B.S." in line or "M.S." in line:
                degree, grad = line.split("Expected Graduation:")
                c.setFont(FONT_BOLD, 11)
                c.drawString(left_margin + 10, y, degree.strip())
                y -= LINE_SPACING
                c.setFont(FONT, 11)
                c.drawString(left_margin + 10, y, "Expected Graduation: " + grad.strip())
                y -= LINE_SPACING
                continue

            if "Relevant Coursework:" in line:
                key, val = line.split(":", 1)
                c.setFont(FONT_BOLD, 11)
                c.drawString(left_margin + 10, y, "Relevant Coursework:")
                y -= LINE_SPACING
                c.setFont(FONT, 11)
                y = draw_wrapped_text(c, val.strip(), left_margin + 20, y, width - 2 * inch)
                y -= 5
                continue

            if ":" in line:
                key, val = line.split(":", 1)
                c.setFont(FONT_BOLD, 11)
                c.drawString(left_margin, y, key.strip() + ":")
                c.setFont(FONT, 11)
                c.drawString(left_margin + FIELD_INDENT, y, val.strip())
                y -= LINE_SPACING
            else:
                y = draw_wrapped_text(c, line, left_margin, y, width - 2 * inch)

            if y < inch:
                c.showPage()
                y = height - inch
                c.setFont(FONT, 11)

        c.save()

print("✅ Polished resumes saved to 'resumes_pdf/' with proper wrapping and formatting.")
```

Step 3: Matching Logic
You’ll create a function to compute:

Role Match (1 or 0)

Research Interest Similarity (Jaccard)

Skill Match (overlap %)

```python
def jaccard_similarity(set1, set2):
    set1, set2 = set(set1), set(set2)
    return len(set1 & set2) / len(set1 | set2) if set1 | set2 else 0

def compute_compatibility(student, professor):
    role_match = 1 if student['desired_role'] in professor['positions_available'] else 0
    interest_match = jaccard_similarity([student['research_interests']], professor['research_interests'])
    skill_match = jaccard_similarity(student['skills'], professor['inferred_skills'])
    score = 0.2 * role_match + 0.4 * interest_match + 0.4 * skill_match
    return round(score * 100, 2)
```

```python
matches = []

for _, student in students_df.iterrows():
    for _, professor in professors_df.iterrows():
        score = compute_compatibility(student, professor)
        matches.append({
            'student_name': student['name'],
            'student_email': student['email'],
            'professor_name': professor['name'],
            'professor_email': professor['email'],
            'desired_role': student['desired_role'],
            'matched_role': professor['positions_available'],
            'compatibility_score': score
        })

matches_df = pd.DataFrame(matches)
matches_df
```

## Actual Gemini API

This code integrates the Gemini Pro LLM API into our Bitcamp project to intelligently infer the technical skills a student should have based on a professor’s research interests. We pass the research topics (e.g., “Smart Cities”, “Cybersecurity”) as a prompt to the Gemini model using the generate_content() method. The model returns a comma-separated list of relevant skills like “Python”, “IoT”, or “Data Analysis”. These inferred skills are then used to calculate a compatibility score between students and professors, enabling smarter matching in our ML-powered recommendation system.

```python
import google.generativeai as genai

# Paste your API key here (from MakerSuite or Bitcamp)
genai.configure(api_key="AIzaSyCR8TmTktb2mV4pJY4_Zt2bgticSHiCGNU")

# Correct model name (DO NOT use gemini-pro without 'models/')
model = genai.GenerativeModel("models/gemini-2.0-flash")
def get_gemini_skills(research_interests):
    prompt = f"""
    Based on the following research interests: {', '.join(research_interests)},
    list 5 technical or programming skills a student should ideally have to work on these topics.
    Only return a comma-separated list of skills.
    """
    try:
        response = model.generate_content(prompt)
        return [skill.strip() for skill in response.text.split(",") if skill.strip()]
    except Exception as e:
        print("Error with Gemini:", e)
        return []
from tqdm import tqdm
import time
import pandas as pd

# Step 1: Gemini for top 10 professors
subset = professors_df.head(10).copy()  # Avoid SettingWithCopyWarning
inferred_skills = []

for interests in tqdm(subset['research_interests']):
    skills = get_gemini_skills(interests)
    inferred_skills.append(skills)
    time.sleep(1.2)  # avoid quota errors

subset['inferred_skills'] = inferred_skills

# Step 2: Skill match score
def skill_match_score(student_skills, prof_skills):
    student_skills, prof_skills = set(student_skills), set(prof_skills)
    if not prof_skills:
        return 0
    return len(student_skills & prof_skills) / len(prof_skills)

# Step 3: Compatibility score function
def compatibility_score(student, professor):
    role_match = 1 if student["desired_role"] in professor["positions_available"] else 0

    student_interest = [student["research_interests"]] if isinstance(student["research_interests"], str) else student["research_interests"]
    research_match = len(set(student_interest) & set(professor["research_interests"])) / \
                     len(set(student_interest) | set(professor["research_interests"]))

    skill_match = skill_match_score(student["skills"], professor["inferred_skills"])

    return round(0.2 * role_match + 0.4 * research_match + 0.4 * skill_match, 2)

# Step 4: Match all students to subset professors
matches = []
for _, student in students_df.iterrows():
    for _, prof in subset.iterrows():
        score = compatibility_score(student, prof)
        matches.append({
            "student": student["name"],
            "professor": prof["name"],
            "score": score,
            "matched_role": prof["positions_available"],
            "student_role": student["desired_role"]
        })

matches_df = pd.DataFrame(matches).sort_values(by="score", ascending=False)

# Step 5: Save matches
matches_df.to_csv("final_matches_subset.csv", index=False)

```
